# Matplotlib and Seaborn
Matplotlib and Seaborn act as the backbone of data visualization through Python. 
 
* Matplotlib: It is a Python library used for plotting graphs with the help of other libraries like Numpy and Pandas. It is a powerful tool for visualizing data in Python. It is used for creating statical inferences and plotting 2D graphs of arrays. It uses Pyplot for providing MATLAB like interface free and open-source. It is capable of dealing with various operating systems and their graphical backends. 
* Seaborn: It is also a Python library used for plotting graphs with the help of Matplotlib, Pandas, and Numpy. It is built on the roof of Matplotlib and is considered as a superset of the Matplotlib library. It helps in visualizing univariate and bivariate data. It uses beautiful themes for decorating Matplotlib graphics. It acts as an important tool in picturing Linear Regression Models. It serves in making graphs of statical Time-Series data. It eliminates the overlapping of graphs and also aids in their beautification.

To use it, you must import the modules:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line chart
A line chart is a type of chart used to show information that changes over time. Line charts are created by plotting a series of several points and connecting them with a straight line

### Matplotlib

```python
# make a list
yields_apples = [0.89,0.91,0.93,0.94,0.95]
```

```python
# plot a list
plt.plot(yields_apples)
```

we can also plot multiple list, see the code below:

```python
# make multiple list
years = [2017,2018,2019,2020,2021,2022]
yields_apples = [0.89,0.91,0.92,0.93,0.94,0.95]
```

```python
# plot multiple list
plt.plot(years, yields_apples)
```

we can change label of x and y, see the code below:

```python
plt.plot(years, yields_apples)
plt.xlabel("year") # change x label
plt.ylabel("Yield (tons per hectare)") # change y label
```

we can also specify the title and legend, also add new line parameters with markers:

```python
# make multiple list
years = [2017,2018,2019,2020,2021,2022]
yields_apples = [0.89,0.91,0.92,0.93,0.94,0.95]
sold = [0.87,0.90,0.88,0.90,0.87,0.89]
```

```python
# plot 2 lines of data
plt.plot(years, yields_apples, marker = 'o') # add marker
plt.plot(years, sold, marker = 'o') # add marker
plt.xlabel("year")
plt.ylabel("value")

plt.title("yields apples and sold per year") # add title
plt.legend(["yields apples", "sold"]) # add legends
```

to configure figure size we can use plt.figure, see the code below:

```python
plt.figure(figsize=(10, 5)) # configure size
plt.plot(years, yields_apples, marker = 'o')
plt.plot(years, sold, marker = 'o')
plt.xlabel("year")
plt.ylabel("value")

plt.title("yields apples and sold per year")
plt.legend(["yields apples", "sold"])
```

### Seaborn

to use seaborn the data must be converted to dataframe

```python
import pandas as pd
```

```python
# make multiple list
years = [2017,2018,2019,2020,2021,2022]
yields_apples = [0.89,0.91,0.92,0.93,0.94,0.95]
sold = [0.87,0.90,0.88,0.90,0.87,0.89]
```

```python
# convert into dataframe
dict_data = dict(years=years, yields_apples=yields_apples, sold=sold)
data = pd.DataFrame(dict_data)
```

here the example of linechart with seaborn:

we can chage background style with sns.set_style

```python
# chage background style
sns.set_style("whitegrid")
```

```python
fig, ax = plt.subplots(figsize=(14, 5)) # configure size
ax = sns.lineplot(x='years', y = 'yields_apples', data = data)# add first line
ax1 = sns.lineplot(x='years', y = 'sold', data = data)# add second line
plt.show()# show chart
```

```python
# chage background style
sns.set_style("darkgrid")
```

```python
fig, ax = plt.subplots(figsize=(14, 5)) # configure size
ax = sns.lineplot(x='years', y = 'yields_apples', data = data)# add first line
ax1 = sns.lineplot(x='years', y = 'sold', data = data)# add second line
plt.show()# show chart
```

## bar chart
bar charts used to compare categories when you have at least one categorical or discrete variable. Each bar represents a summary value for one discrete level, where longer bars indicate higher values. 

### Matplotlib

Here a simple barchart with a few values:

```python
plt.figure(figsize=(10, 5))
plt.bar(years, yields_apples)
plt.xlabel("year")
plt.ylabel("value")

plt.title("yields apples per year")
```

we can add more values and make it to staked barchart:

```python
plt.figure(figsize=(10, 5))
plt.bar(years, yields_apples)
plt.bar(years, sold)
plt.xlabel("year")
plt.ylabel("value")

plt.title("yields apples and sold per year")
plt.legend(["yields apples", "sold"])
```
